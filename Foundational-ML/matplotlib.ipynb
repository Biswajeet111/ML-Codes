```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Name" : ["Aman","Riya","Rahul","Sneha","Arjun","Priya","Karan","Neha"],\
    "Age" : [22,25,24,23,26,27,28,22],
    "Gender" : ["Male","Female","Male","Female","Male","Female","Male","Female"],
    "Marks" : [75,88,92,70,85,90,78,95]
}

df = pd.DataFrame(data)
df
```

### Line Plot for Marks Trends

```python
plt.plot(df["Marks"])

plt.title("Marks Trend")
plt.xlabel("Students")
plt.ylabel("Marks")
```

## Bar plot for comparison

```python
plt.bar(df["Name"],df["Marks"])

plt.title("Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")
```

## Histogram for distribution of Age

#### Bin Width = Range/bin

```python
plt.hist(df["Age"],bins=5)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

#### Scatter Plot

Relationship between Age and Marks

```python
plt.scatter(df["Age"],df["Marks"])

plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
```

#### Pie Chart

Gender Distribution

```python
gender_counts = df["Gender"].value_counts()

plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%")

plt.title("Gender Distribution")
```

#### Customizing Plots

Example with colors and grid

```python
plt.bar(df["Name"],df["Marks"],color="purple")

plt.title("Student Marks")
plt.xlabel("Student Name")
plt.ylabel("Marks")

plt.grid(True)
```

### SubPlots

Show two graphs together

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(df["Age"])
plt.title("Age Distribution")

plt.subplot(1,2,2)
plt.bar(df["Name"],df["Marks"])
plt.title("Marks of Students")
```
