## GroupBy in pandas

```python
data = {
    "Name" : ["Aman","Saurav","Karan","Subham","Abhishek",None,"Sameer"],
    "Age" : [22,23,None,22,25,24,26],
    "Marks" : [85,None,88,90,56,72,None],
    "Subject" : ["Math","Physics","Math","Chemistry","Math","Physics","Chemistry"]
}

df=pd.DataFrame(data)

print(df)
```

```python
#basic groupby
df.groupby('Subject')['Marks'].mean()  ## Mean marks per subject
```

```python
df.groupby('Subject')['Name'].count()  ## Count Students per Subject
```

```python
df.groupby('Subject').size()
```

```python
##Multiple Aggregation
df.groupby('Subject')['Marks'].agg(['mean','max','min'])
```

```python
df.groupby('Subject')['Marks'].agg(['mean','max','min','count']) ## adding count here also
```

```python
##Groupby by multiple columns
df.groupby(['Subject','Age'])['Marks'].mean()
```
