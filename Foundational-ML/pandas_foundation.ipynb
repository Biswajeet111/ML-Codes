## DataFrame in pandas

```python
data={
    "Name": ["Raju","Bheem","Chutki"],
    "Age": [9,24,17],
    "Marks": [65,90,69]
}
```

```python
print(data)
```

```python
type(data)
```

```python
df=pd.DataFrame(data)
```

```python
print(df) ##Pandas automatically adds a default index: this is called a range Index(Start from 0 by default)
```

```python
#we can also add custom index in pandas dataframe
df= pd.DataFrame(
    data,
    index =["Student 1","Student 2","Student 3"]
)
```

```python
print(df)
```

```python
df.set_index("Name",inplace=True) #if we are setting in place = true it means we are changing the original object it self. if we do this without inplace it will create a new dataframe
```

```python
print(df)
```

### Handling Missing values

```python
##Updating the dataframe with some null values
df=pd.DataFrame({
    "Name": ["Raju","Bheem","Chutki","Kalia",None,"Indumati"],
    "Age": [9,24,17,20,30,None],
    "Marks": [65,None,69,85,78,None]
})

df
```

```python
df.isnull() ##to check for null values in the dataframe
```

```python
df.isnull().sum() ##to count the number of null values in each column
```

```python
df.dropna()
```

```python
print(df)
```

```python
clean_df=df.dropna() ##to create a new dataframe without null values
```

```python
print(clean_df)
```
