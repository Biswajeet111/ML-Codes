```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Name": ["Gaurav", "Sharan", None, "Rajesh", "Priya", None, "Vikram", "Anjali", "Arjun", None],
    "Department": ["Sales", None, "IT", "HR", None, "Finance", "Sales", "IT", None, "HR"],
    "Role": ["Manager", "Executive", None, "Lead", "Developer", "Analyst", None, "Designer", "Engineer", "Coordinator"],
    "Experience": [5, 3, None, 7, 2, 4, None, 6, 8, 1],
    "Salary": [50000, 35000, None, 60000, 40000, 45000, None, 48000, 55000, 32000],
    "Score": [8.5, 7.2, None, 9.1, 6.8, 7.9, None, 8.3, 9.0, 6.5]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.info()
```

What is average mean salary for each department

What is average mean performance score for each role



```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Role')['Score'].mean()
```

Counting and sizing

how many actual non null performance score entries exist for each department

total number of employee records(rows)in each department,regardless of whether data is missing

```python
df.groupby('Department')['Score'].count()
```

```python
df.groupby('Department').count()
```

```python
df.groupby('Department').size()
```

Multiple Aggregation

```python
df.groupby('Department')['Salary'].agg(['mean','max','min','count'])
```

```python
df.groupby('Role')['Score'].agg(['mean','max','min','count'])
```

Grouping by multiple columns

find avg salary of both grouped department and role

find avg score grouped by both department and experiene level

```python
df.groupby(['Department','Role'])['Salary'].mean()
```

```python
df.groupby(['Department','Experience'])['Score'].mean()
```
