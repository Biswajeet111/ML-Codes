# Case Study 
**A university monitors electricity usage (in units) of 4 hostels over 7 days(Mon-Sun)**
##
**Array-Name:Electricity**
##
**Shape(4,7)**
##
**Rows:Hostels(H1,H2,H3,H4)**
##
**Cols:Days(Mon-Sun)**

```python
import numpy as np
```

# Task A1

```python
arr=np.random.randint(120,650,(4,7))
```

```python
print(arr)
```

# Task A2

```python
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
```

# Task B1

```python
avg_usage=np.mean(arr,axis=1)
max_usage=np.max(arr,axis=1)
highest_avg=np.max(avg_usage)
```

```python
print(avg_usage)
print(max_usage)
print(highest_avg)
```

# Task B2

```python
total_consumption=np.sum(arr,axis=0)
highest_total=np.max(total_consumption)
```

```python
print(total_consumption)
print(highest_total)
```
