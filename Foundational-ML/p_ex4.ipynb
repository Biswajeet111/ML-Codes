```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

### Data Creation

```python
data = {
"Exercice Hrs": [1,-2, 3, 1, 0, 2, 3, 1, 0, 2, 1, 3, 2, 1, 0 , 2, 3, 1, 2, 1],
"Calories In" : [2000, 2200, 2500, 2100, 1800, 2300, 2600, 2000, 1900, 2200, 2100, 2400, 2300, 2000, 1800, 2200, 2500, 2100, 2300, 2000],
"Spotify Hrs" : [1, 2, 2, 1, 1, 3, 2, 1, 2, 2, 1, 2, 3, 1, 1, 2, 2, 1, 2, 1 ],
"Step a Day": [5000, 7000, 8000, 6000, 4000, 8000, 10000, 5500, 4500, 7200, 6000, 8500, 9200, 5800, 4000, 7600, 9500, 6100, 8000, 690],
"Heart Rate": [72, 75, 75, 74, 70, 76, 80, 73, 71, 75, 74, 77, 79, 73, 70, 76, 81, 74, 77, 73]
}

df = pd.DataFrame(data)
```

### Statistical Analysis

```python
df.describe().T
```

### Visualizations

Scatter Plot of (Excercise Vs HeartRate)

```python
plt.scatter(df["Exercice Hrs"],df["Heart Rate"])
plt.title("Excersise Vs HeartRate")
plt.xlabel("Excersise")
plt.ylabel("HeartRate")
```

Histogram(Steps)

```python
plt.hist(df["Step a Day"],bins=5)

plt.title("Steps Count")
plt.xlabel("No of Steps")
plt.ylabel("Frequency")
```

Bar Chart(Calories)

```python
plt.bar(df["Exercice Hrs"],df["Calories In"])

plt.title("Calories Burnout")
plt.xlabel("Hrs of Exercise")
plt.ylabel("Calories")
```
