```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()

X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)

X_train_scaled.shape, X_test_scaled.shape
```

`.fit()` ---> calculate mean and std deviation from training only

`transform()` --->applies learned parameters

## Test data never influences training statistics

# Low Variance Threshold

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
selector = VarianceThreshold(0.01)
X_train_var = selector.fit_transform(X_train_scaled)
X_test_var = selector.transform(X_test_scaled)
X_train_var.shape , X_train_scaled.shape
```

No low variance Features

# Correlation based Feature Removal 

This dataset is famous for multicollinearity

```python
corr_matrix = pd.DataFrame(X_train_scaled, columns= X.columns).corr().abs()
```

```python
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k = 1).astype(bool))
to_drop = [col for col in upper.columns if any(upper[col]>0.9)]
to_drop
```

```python
X_train_corr = pd.DataFrame(X_train_scaled,columns= X.columns).drop(columns= to_drop)
X_test_corr = pd.DataFrame(X_test_scaled , columns= X.columns).drop(columns= to_drop)
```

```python
X_test_corr.shape, X_train_corr.shape
```

# Tree based Feature Importance

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state=42)
```

```python
rf.fit(X_train_corr, y_train)
```

```python
importances = pd.Series(rf.feature_importances_, index = X_train_corr.columns)
importances.sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(8,6))
importances.plot(kind='barh')
plt.gca().invert_yaxis()
plt.title("Feature Importances (All 20 Features)")
plt.xlabel("Importance")
plt.show()
```

### PCA (Principal Component Analysis)


Only After Coorelation and is used for doing the Dimensionality Reduction

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X_train_corr)

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.show()
```

```python
pca = PCA(n_components = 0.95)
X_train_pca = pca.fit_transform(X_train_corr)
X_test_pca = pca.transform(X_test_corr)

X_train_pca.shape
```
